```python
from src.ANN import *
import numpy as np
from src.dump import dump, load
from src.encoder import *
```

```python
model = Sequential()
model.add(Dense(units=2, input_dim=4, activation_function="relu"))
model.add(Dense(units=16, activation_function="relu"))
model.add(Dense(units=16, activation_function="relu"))
model.add(Dense(units=3, activation_function="relu"))

layer_softmax = SoftMax()
model.add(layer_softmax)
```

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df_x = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
df_y = df[['Species']]
```

```python
df_y_encoded = encode(df_y.values)
```

```python
df_x_values = df_x.values
```

```python
model.compile("cross_entropy_error", 0.01, 0.01)
model.fit(df_x_values, df_y_encoded, 10, 100)
```

```python
# Dump model
dump(model, "softmax2_model.json")

# Load data model
loaded_data = load("softmax2_model.json")

loaded_model = Sequential()
loaded_model.useJSON(loaded_data)
```

```python
loaded_model.predict(df_x_values)
```
